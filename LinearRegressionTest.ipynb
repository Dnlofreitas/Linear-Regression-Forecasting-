```python
import pandas as pd
import numpy as np

data_regression = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 5, 7, 11]
}

data_ar = {
    'y': [2, 3, 5, 7, 11]
}

df_regression = pd.DataFrame(data_regression)
df_ar = pd.DataFrame(data_ar)

with pd.ExcelWriter('output.xlsx') as writer:
    df_regression.to_excel(writer, sheet_name='Regression Linear Simple', index=False)
    df_ar.to_excel(writer, sheet_name='Autoregressive Model', index=False)

print("Data saved to output.xlsx")
```

```python
from statsmodels.tsa.ar_model import AutoReg
```

```python
print(data_ar)
```

```python
model = AutoReg(df_ar['y'], lags=1)
```

```python
model_fit = model.fit()
```

```python
predictions = model_fit.predict(start=len(df_ar), end=len(df_ar)+1)
print(predictions)
```
